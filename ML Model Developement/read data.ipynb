```python
category_df = {}
opex_main_categories = opex_data['dmk_name'].unique()
for x in opex_main_categories:
    category_df[f'{x}'] = opex_data[opex_data['dmk_name'] == x]
    print(f'{x} dataframe created.')
```

```python
category_name = {}

for x in opex_main_categories:
    sub_df = category_df[f'{x}']
    sub_categories = sub_df['dsk_name'].unique()
    category_name[f'{x}'] = sub_categories  
    
```
